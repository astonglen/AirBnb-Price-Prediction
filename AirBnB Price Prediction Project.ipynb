### **ABSTRACT**

<font size="3">
The ability to predict prices and features affecting the appraisal of property can be a powerful tool in such a cash intensive market for a lessor. Additionally, a predictor that forecasts the number of reviews a specific listing will get may be helpful in examining elements that affect a property's popularity. The issue has two components: establishing what factors affect the price of a listing and customer happiness for the approximately 50000 Airbnb properties listed in New York City, as well as investigating what factors affect the number of reviews and customer popularity of the rentals. We need to create models that can confidently forecast the price and amount of reviews of listed homes in order to solve this issue.
    </font>

### **INTRODUCTION**

<font size="3">
Airbnb is a peer-to-peer online platform that enables hosts with extra properties to rent them out to guests looking for lodging in that area. Setting an ideal price for their listings on Airbnb is crucial for hosts because it boosts client attraction, contentment, customer flow, and host revenue. All of them are a direct result of better supply and demand matching, which also has an effect on the host's turnover.
</font>

<font size="3">
The number of reviews a property listing receives is another element that may be utilized to assess a property's potential popularity as well as to investigate and examine the variables that lead to less reviews for particular properties.
This can be helpful in figuring out how to increase the typical number of reviews that real estate listings receive, which will help the hosts increase their revenue and make the platform more competitive with its rivals.
</font>

<font size="4">**B. Model Training**</font>

<font size="3">The dataset is split into 70-30 for training and testing using sklearn. Each model’s effectiveness is compared using their R2 values and Root Mean Square Errors (RMSE). </font>


<font size="3"> The following regression models are used:<br><br></font>
<font size="3">1. Linear Regression<br>
&emsp;This model optimizes the linear combination of the predictors, but most real world problems are unfortunately not so simple because of &emsp;the highly interconnected nature of the observed variables. <br><br>
2. Bayesian Ridge regression<br>
&emsp;It adds a penalizing term to the squared error cost function. It has been noted to perform well in datasets which were observed to have &emsp;high variance and reduce overfitting, but is known to converge well only for linearly separable data.<br><br>
3. Stochastic Gradient Descent (SGD)<br>
&emsp;It uses SGD to determine the coefficients. It is a very simple yet efficient approach to minimizing the cost functions but is limited to &emsp;&emsp;&ensp;&nbsp;&nbsp;&nbsp;capturing relationships ranging from simple to moderate complexities.<br><br>
4. Lasso Regression<br>
&emsp;It is a type of linear regression that uses shrinkage, i.e., where the values are shrunk towards a central point. It encourages simple &emsp;&emsp;&ensp;&nbsp;&nbsp;&nbsp;models determined by fewer parameters by placing penalties on attributes with low correlation.<br><br>
5. Random Forest Regression<br>
&emsp;It is from the family of decision trees. Has a high accuracy, and the ability to handle large number of features, tuning the &emsp;&emsp;&ensp;&nbsp;&nbsp;&nbsp;&nbsp;hyperparameters has always been a challenge.</font>

```python
#comparing the R2 value for all the methods with number of components

from sklearn.decomposition import PCA
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

from sklearn import linear_model
from sklearn.linear_model import Lasso

scores=[]
lasso_scores=[]
ridge=[]
RF1=[]
for i in range(1,200,10):
    
    pca0=PCA(n_components=i).fit_transform(X)
    
    pca0=pd.DataFrame(pca0)
    
    ny_train,ny_test,price_train,price_test = train_test_split(pca0,y,test_size=0.3)
    ny_lr=LinearRegression().fit(ny_train,price_train)
    ny_regL1 = Lasso(alpha=0.01).fit(ny_train,price_train)
    ny_ridge=linear_model.BayesianRidge().fit(ny_train,price_train)

    predicted_val_lr=ny_lr.predict(ny_test)
    predicted_val_lasso=ny_regL1.predict(ny_test)
    predicted_val_ridge=ny_ridge.predict(ny_test)

    x_lr=r2_score(price_test,predicted_val_lr)
    x_lasso=r2_score(price_test,predicted_val_lasso)
    x_ridge=r2_score(price_test,predicted_val_ridge)
    
    scores.append(x_lr)
    lasso_scores.append(x_lasso)
    ridge.append(x_ridge)

x = np.arange(1,200,10)

# plotting

plt.title("LR")
plt.plot(x,scores,'y-')
plt.xlabel("components")
plt.ylabel("R2 scores on test set ")
plt.show()

plt.title("Lasso")
plt.plot(x,lasso_scores,'g-')
plt.xlabel("components")
plt.ylabel("R2 scores on test set ")
plt.show()

plt.title("Ridge")
plt.plot(x,ridge,'b-')
plt.xlabel("components")
plt.ylabel("R2 scores on test set ")
plt.show()
```

### **From the above graphs , we observe an elbow point for 40 components.Further applying Random Forest with N_components as 40 we get the following results**

```python
pca0=PCA(n_components=40).fit_transform(X)

pca0=pd.DataFrame(pca0)

ny_train,ny_test,price_train,price_test = train_test_split(pca0,y,test_size=0.3)

regrRM = RandomForestRegressor(n_estimators=300).fit(ny_train,price_train)

pred_val_RF1=regrRM.predict(ny_test)

x_RF2 = regrRM.score(ny_train,price_train)
rmse = np.sqrt(metrics.mean_squared_error(price_test, pred_val_RF1))
print(x_RF2)
print('RMSE:',rmse)
```
